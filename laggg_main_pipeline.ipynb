```python
#координаты p = p[x1:x2,y1:y2] ----> (x1,x2,y1,y2) 
hp = (697,700,507,777) #p[697:700,507:777]
sp = (673,676,507,777) #p[673:676,507:777]

bandage = (130,158,1215,1243) #p[130:158,1215:1243]
medicine = (157,185,1215,1243) #p[157:185,1215:1243]
cola = (185,213,1215,1243) #p[185:213,1215:1243]
pills = (212,240,1215,1243) #p[212:240,1215:1243]

yellow = (270,298,1245,1273) #p[270:298,1245:1273]
red = (315,343,1245,1273) #p[315:343,1245:1273]
blue = (355,383,1245,1273) #p[355:383,1245:1273]
green = (400,428,1245,1273) #p[400:428,1245:1273]

right_bullets = (642,670,685,713) #p[642:670,685:713]
left_bullets = (636,664,610,670) #p[636:664,610:670]
zoom = (15,43,610,740) #p[15:43,610:740]

helmet = (650,678,890,918) #p_bgr[650:678,890:918,2]
vest = (650,678,930,958) #p_bgr[650:678,930:958,2]
backpack = (650,678,970,998) #p_bgr[650:678,970:998,2]
all_kills = (541,569,605,633) #p_bgr[541:569,605:633,2]

aw1 = (525,550,1240,1270)
aw2 = (575,600,1240,1270)
aw3 = (625,650,1240,1270)
aw4 = (675,700,1240,1270)

door = (407,427,670,710) #p[407:427,670:710]
action = (290,310,640,680) #p[290:310,640:680] #для браузера p[310:330,640:680]

w1 = (550,575,1200,1270) #[550:575,1200:]
w2 = (600,625,1200,1270) #[600:625,1200:]
w3 = (650,675,1200,1270) #[650:675,1200:]
w4 = (700,717,1200,1270) #[700:,1200:]
#--------------------------------------------------------------------------------

all_numbs = [bandage,medicine,cola,pills,
              yellow,red,blue,green,
              helmet,vest,backpack,
              right_bullets,left_bullets,
              all_kills, zoom]
active_weapons = [aw1, aw2, aw3, aw4]
weapons = [w1,w2,w3,w4]
```

```python
%%time

cap = cv2.VideoCapture("videoplayback.mp4")
picts = []
ts_map = []
ts_hp = []
ts_sp = []
ts_dir = []
ts_numb = []
ts_actweap = []
ts_hand = []
ts_door = []
ts_act = []
ts_weap = []
ts_cursor = []


i = 0
while True:
    ret,frame=cap.read()  #frame.shape=718x1280x3
    if not ret:  # or i == 28*30
        break
    gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #-----------------------------
    if open_map(frame):
        ts_map.append(i)
    ts_hp.append(detect_hp(gray_img, hp))
    ts_sp.append(detect_sp(frame, sp))
    ts_dir.append(find_centers(gray_img))
    ts_numb.append(identify_all_numbers(frame, model_numb, all_numbs))
    actw = active_weapon(gray_img, active_weapons)
    ts_actweap.append(actw)
    if actw == 3:
        ts_hand.append(r_hands(gray_img))
    else:
        ts_hand.append(0.0)    
    ts_door.append(open_close_door(gray_img, door))
    ts_act.append(detect_using_reload(gray_img, action))
    ts_weap.append(all_weapons(gray_img, weapons))
    #-----------------------------
    if i%28==0: # каждую секунду
        aim_xy = where_aim(gray_img, aim)
        prev_x = aim_xy[0][0]
        prev_y = aim_xy[1][0]
        now_x = aim_xy[0]
        now_y = aim_xy[1]
        now_x = now_x[(now_x - prev_x).argmin()] # берем самый близкий к курсору из предыдущего кадра
        now_y = now_y[(now_y - prev_y).argmin()] # берем самый близкий к курсору из предыдущего кадра
    elif i%3==0:
        part_i1 = max(prev_x-aim_window,0)
        part_i2 = prev_x+aim_window
        part_j1 = max(prev_y-aim_window,0)
        part_j2 = prev_y+aim_window
        p1 = gray_img[part_i1:part_i2,part_j1:part_j2]
        aim_xy = where_aim(p1, aim)
        now_x = aim_xy[0]+max(prev_x-aim_window,0)
        now_y = aim_xy[1]+max(prev_y-aim_window,0)
        now_x = now_x[(now_x - prev_x).argmin()] # берем самый близкий к курсору из предыдущего кадра
        now_y = now_y[(now_y - prev_y).argmin()] # берем самый близкий к курсору из предыдущего кадра
    else:
        now_x = prev_x
        now_y = prev_y
    prev_x = now_x
    prev_y = now_y
    ts_cursor.append([now_x, now_y])
    #-------------------------------------------------------
    i += 1
    picts.append(gray_img)


ts_map = np.array(ts_map)
ts_hp = np.array(ts_hp)
ts_sp = np.array(ts_sp)
ts_numb = np.array(ts_numb)
ts_actweap = np.array(ts_actweap)
ts_hand = np.array(ts_hand)
ts_door = np.array(ts_door)
ts_act = np.array(ts_act)
ts_weap = np.array(ts_weap)
ts_cursor = np.array(ts_cursor)
```
